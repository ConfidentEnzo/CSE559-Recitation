```python
import numpy as np
from skimage.io import imread, imsave
from scipy.signal import convolve2d as conv2
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# Load an image and add noise
img = np.float32(imread('example.jpg'))/255.
img = img + np.random.normal(scale=0.2,size=img.shape)
plt.imshow(img, cmap='gray')
```

```python
# Define a simple smoothing kernel
k = np.ones([9,9])
print(k)
```

```python
# Smooth the image with kernel k and normalize.
# But for locations outside the boudary, set the
# contribution to 0 and normalize accordingly.

# Convolve image with the kernel
output = conv2(img, k, mode='same', boundary='fill', fillvalue=0)

# Convolve a all-one matrix with the kernel 
# to get the normalizing factor for each output pixel 
ones = np.ones_like(img)
norm = conv2(ones, k, mode='same', boundary='fill', fillvalue=0)

# Normalize
output = output / norm
# output = output / 81


plt.imshow(output, cmap='gray')
```
